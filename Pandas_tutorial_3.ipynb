```python
import numpy as np
import pandas as pd
```

```python
inventaris = pd.read_csv('toko_inventaris.csv', delimiter = ',',names = ['barang','merek','jumlah'])
```

```python
restock = pd.read_csv('toko_restock.csv', delimiter = ',',names = ['barang','restock sebulan'])
```

```python
inventaris.merge(restock)
```

```python
inventaris.merge(restock, how = 'outer')
```

```python
inventaris.merge(restock, how = 'left')
```

```python
inventaris.sort_values('jumlah')
```

```python
inventaris.sort_values(by = ['barang','jumlah'])
```

```python
inventaris.sort_values('jumlah', ascending = False)
```

```python
inventaris.sort_values('jumlah', ascending = False).reset_index()
```

```python
inventaris.sort_values('jumlah', ascending = False).reset_index(drop = True)
```

```python
inventaris.merge(restock, how = 'left').fillna('unknown')
```

```python
inventaris.merge(restock, how = 'left').fillna('4')
```

```python
article_read = pd.read_csv('pandas_tutorial_read.csv', delimiter=';', names = ['my_datetime', 'event', 'country', 'user_id', 'source', 'topic'])
```

```python
blog_buy = pd.read_csv('pandas_tutorial_buy.csv', delimiter=';', names = ['my_date_time', 'event', 'user_id', 'amount'])
```

```python
data_merged = article_read.merge(blog_buy, how = 'left', left_on = 'user_id', right_on = 'user_id')
```

```python
article_read.merge(blog_buy, how = 'left', left_on = 'user_id', right_on = 'user_id').amount.fillna(0).mean()
```

```python
data_merged.fillna(0).groupby('country').sum().amount.sort_values( ascending = False).head(3)
```
